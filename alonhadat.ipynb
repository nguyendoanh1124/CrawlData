```python
import httpx
from bs4 import BeautifulSoup
import json
```

```python
def get_html_from_url(url):
    with httpx.Client(http2=True) as client:
        response = client.get(url=url)
        return response.text
```

```python
url_website = get_html_from_url('https://alonhadat.com.vn/du-an-can-ho-chung-cu-sun-cosmo-residence-da-nang-pj10694')
print(url_website)
```

```python
def extract_html_info(html):
    soup = BeautifulSoup(html, 'lxml')

    title = soup.select_one('.content-left .project-name h1').text
    return {
        'title': title.strip(),
    }
```

```python
extract_info = extract_html_info(url_website)
print(extract_info)
```
